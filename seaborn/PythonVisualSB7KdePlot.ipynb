```python
"""
seaborn系列 (7) | 核函数密度估计图kdeplot() 原创
脑机接口社区2022-01-24 15:58:40博主文章分类：数据分析与可视化

核函数密度估计图
核函数密度估计图kdeplot(),该图主要用来拟合并绘制单变量或双变量核密度估计值。

函数原型
seaborn.kdeplot(data, data2=None, shade=False, 
                vertical=False, kernel='gau', bw='scott',
                gridsize=100, cut=3, clip=None,
                legend=True, cumulative=False,
                shade_lowest=True, cbar=False, 
                cbar_ax=None, cbar_kws=None, ax=None, **kwargs)

参数解读
shade:阴影：bool类型
作用：设置曲线下方是否添加阴影，如果为True则在曲线下方添加阴影
(如果数据为双变量则使用填充的轮廓绘制)，若为False则，不绘制阴影

cbar:bool类型
作用：如果为True则绘制双变量KDE图，并添加颜色条
-----------------------------------
©著作权归作者所有：来自51CTO博客作者脑机接口社区的原创作品，请联系作者获取转载授权，否则将追究法律责任
seaborn系列 (7) | 核函数密度估计图kdeplot()
https://blog.51cto.com/u_6811786/4967109
"""
import matplotlib.pyplot as plt
import numpy as np; 
import seaborn as sns; 
import warnings
warnings.filterwarnings("ignore")
sns.set(color_codes=True)
np.random.seed(10)
#构建数据
mean, cov = [0, 2], [(1, .5), (.5, 1)]
x, y = np.random.multivariate_normal(mean, cov, size=50).T
```

```python
"""
案例1：
绘制基本的单变量密度曲线图
"""
sns.kdeplot(x)
plt.show()
```

```python
"""
案例2：
绘制密度曲线，并在曲线下着色
通过设置color来设置不同的显示颜色
"""
sns.kdeplot(x, shade=True, color="g")
plt.show()
```

```python
"""
案例3：
绘制双变量密度曲线
"""
sns.kdeplot(x=x, y=y)
plt.show()
```

```python
"""
案例 4:
使用轮廓填充
"""
sns.kdeplot(x=x, y=y, shade=True,color='r')
plt.show()
```

```python
"""
案例5：
使用更多轮廓级别和不同的调色板：
"""
sns.kdeplot(x=x, y=y, n_levels=30, cmap="Purples_d")
plt.show()
```

```python
"""
案例6：
给轮廓添加颜色条
"""
sns.kdeplot(x=x, y=y, cbar=True)
plt.show()
```

```python
"""
案例7：
绘制多个阴影双变量密度图(这里绘制两个)
"""
# 获取数据
iris = sns.load_dataset("iris", data_home='./seaborn-data', cache=True)
setosa = iris.loc[iris.species == "setosa"]
virginica = iris.loc[iris.species == "virginica"]
sns.kdeplot(x=setosa.sepal_width, y=setosa.sepal_length,cmap="Reds", shade=True, shade_lowest=False)

sns.kdeplot(x=virginica.sepal_width, y=virginica.sepal_length,cmap="Blues", shade=True, shade_lowest=False)
plt.show()
```
